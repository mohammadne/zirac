```python
# lambda functions

salary_list = [100000, 200000, 150000, 120000, 80000, 750000]

# Calculate total_salary of base_salary * (1 + bonus_rate)
def calculate_salary(base_salary: int, bonus_rate: float=.1) -> float:
  return base_salary * (1 + bonus_rate)

total_salary_list1 = [calculate_salary(salary) for salary in salary_list]
print(total_salary_list1)

total_salary_list2 = [(lambda x: x * (1.1))(salary) for salary in salary_list]
print(total_salary_list2)
```

```python
class Mammal:
    def info(self):
        print("Mammal info")


class WingedAnimal:
    def info(self):
        print("WingedAnimal info")


# MRO is 1.Mammal and 2.WingedAnimal
class Bat(Mammal, WingedAnimal):
    def info(self):
        super().info()
        WingedAnimal.info(self)
        print("Bat info")


b = Bat().info()
```

Python creates a function object called info inside the class WingedAnimal
At this point, it’s just a function; it does not automatically know which instance it belongs to.
Methods normally receive the instance (self)
When you call a method on an instance:

w = WingedAnimal()
w.info()

Python internally does this:

WingedAnimal.info(w)

It passes the instance w explicitly as the first argument (self).
w.info() is syntactic sugar for WingedAnimal.info(w).
So w.info() is not instantiation, it’s binding the instance to the method.

```python
from __future__ import annotations


class Point:
    def __init__(self, x: int = 0, y: int = 0):
        self.x = x
        self.y = y

    def __add__(self, other: Point) -> Point:
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)

p1 = Point(1, 2)
p2 = Point(2, 3)

# this statment calls the __add__() method
p3 = p1 + p2

print((p3.x, p3.y))   # Output: (3, 5)
```
